```python
file = uproot.open('DAOD_PHYSLITE.38191914._000036.pool.root (1).1')

tree = file['CollectionTree']

#Branch'leri görelim

print(f"Total branches: {len(tree.keys())}")
```

```python
n_events = tree.num_entries
print(f"Total events in tree: {n_events}")
```

```python
tree.show()
```

```python
all_branches = tree.keys()

photon_branches=[]
for branch in all_branches:
    if 'photon' in branch.lower():
        photon_branches.append(branch)

print(f"\n Toplam photon branch sayısı: {len(photon_branches)}")

print("\nPHOTON BRANCHLERİ:\n")
for branch in photon_branches:
    print(branch)

```

 Truth (AnalysisPhotons içinde):

```python
pt = tree["AnalysisPhotonsAuxDyn.pt"].array(entry_stop=5)
eta = tree["AnalysisPhotonsAuxDyn.eta"].array(entry_stop=5)
phi = tree["AnalysisPhotonsAuxDyn.phi"].array(entry_stop=5)
mass = tree["AnalysisPhotonsAuxDyn.m"].array(entry_stop=5)

ptcone20 = tree["AnalysisPhotonsAuxDyn.ptcone20_CloseByCorr"].array(entry_stop=5)
topoetcone20 = tree["AnalysisPhotonsAuxDyn.topoetcone20_CloseByCorr"].array(entry_stop=5)
topoetcone40 = tree["AnalysisPhotonsAuxDyn.topoetcone40_CloseByCorr"].array(entry_stop=5)

f1 = tree["AnalysisPhotonsAuxDyn.f1"].array(entry_stop=5)

cleaning = tree["AnalysisPhotonsAuxDyn.DFCommonPhotonsCleaning"].array(entry_stop=5)
isEMTight = tree["AnalysisPhotonsAuxDyn.DFCommonPhotonsIsEMTight"].array(entry_stop=5)

OQ = tree["AnalysisPhotonsAuxDyn.OQ"].array(entry_stop=5)

truthType = tree["AnalysisPhotonsAuxDyn.truthType"].array(entry_stop=5)
truthOrigin = tree["AnalysisPhotonsAuxDyn.truthOrigin"]. array(entry_stop=5)


```

```python
for i in range(5):
    print(f"EVENT {i}")
    print(f"Foton Sayısı: {len(pt[i])}")

    
    print(f"pt (Transverse Momentum, MeV):")
    print(f"  {pt[i]}")
   
    
    print(f"eta (Pseudorapidity):")
    print(f"  {eta[i]}")

    
    print(f"phi (Azimuthal Angle):")
    print(f"  {phi[i]}")
 
    
    print(f"mass (MeV):")
    print(f"  {mass[i]}")
  
    
    print(f"ptcone20_CloseByCorr (MeV):")
    print(f"  {ptcone20[i]}")
   
    
    print(f"topoetcone20_CloseByCorr (MeV):")
    print(f"  {topoetcone20[i]}")
  
    
    print(f"topoetcone40_CloseByCorr (MeV):")
    print(f"  {topoetcone40[i]}")

    
    print(f"f1 (Shower Shape):")
    print(f"  {f1[i]}")

    
    print(f"DFCommonPhotonsCleaning:")
    print(f"  {cleaning[i]}")
    
    
    print(f"DFCommonPhotonsIsEMTight:")
    print(f"  {isEMTight[i]}")
    
    
    print(f"OQ (Object Quality):")
    print(f"  {OQ[i]}")
    
    
    print(f"truthType:")
    print(f"  {truthType[i]}")
    
    
    print(f"truthOrigin:")
    print(f"  {truthOrigin[i]}")

    
```
